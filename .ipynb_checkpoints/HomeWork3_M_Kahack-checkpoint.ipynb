## Problem 1

Load the *Online  Retaildataset* (**Online Retail.xlsx**) from the **UCI Machine Learning Repository** into Python using a Pandas dataframe.  Using the apriori module from the **MLxtend** library,  generate Frequent Itemsets for all trans-actions for the country of France.  What itemset has the largest support?  Set the minimum support threshold to 5% and extract frequent itemsets, and use them as input to the association rules module.  Use each of the confidence and lift metrics to extract the association rules with the highest values, respectively. What are the antecedents and consequents of each rule?  Is the rule with the highest confidence the same as the rule with the highest lift?  Why or why not?


### Answer :-
The itemset with the largest support is {'RABBIT NIGHT LIGHT'}. 

The rule with the highest confidence is {'SET/20 RED RETROSPOT PAPER NAPKINS','SET/6 RED SPOTTY PAPER PLATES'} -> {'SET/6 RED SPOTTY PAPER CUPS'}. 

The rule with the highest lift is {'PACK OF 6 SKULL PAPER CUPS'} -> {'PACK OF 6 SKULL PAPER PLATES'}. 

The rule with the highest confidence and the rule with the highest lift are different. This because the lift takes into account the joint support of the rule and both the support of the antecedents and consequents support. Confidence only takes into account joint support of the rule and the antecedent's support.

```python
df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df.head()
```

```python
df.describe()
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis = 0, subset=['InvoiceNo'], inplace = True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
df.describe()
```

```python
france_df = df[df["Country"] == "France"]
france_df.describe()
```

```python
french_basket = (france_df
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

french_basket.tail()
```

```python
def sum_to_boolean(x):
    if x<=0:
        return 0
    else:
        return 1

french_basket_sg = french_basket.applymap(sum_to_boolean)
```

```python
french_basket_sg.drop('POSTAGE', inplace=True, axis=1)
french_basket_sg.describe()
```

```python
frequent_itemsets = apriori(french_basket_sg, min_support = 0.05, use_colnames = True)

frequent_itemsets.sort_values('support', ascending = False).head()
```

```python
frequent_itemsets.sort_values('support', ascending = False).tail()
```

```python
con_rules = association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
con_rules.sort_values('confidence',ascending = False)
```

```python
lift_rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1)
lift_rules.sort_values('lift',ascending = False).head()
```

## Problem 2

Load the *Extended Bakery* dataset (**75000-out2-binary.csv**) into Python using a Pandas dataframe.  Calculate the binary correlation coefficient Φ for the *Chocolate  Coffee* and *Chocolate  Cake* items.   Are  these  two  items  symmetric binary variables?  Provide supporting calculations.  Would the association rules {**Chocolate Coffee**} ⇒ {**Chocolate Cake**} have  the  same  value  for  Φ  as {**Chocolate Cake**} ⇒ {**Chocolate Coffee**}?

### Answer :-

The binary correlation is 0.4856. These items are assymmetric binary variables. <br>

s({Chocolate Cake}) = 0.0835 <br>
s({Chocolate Coffee}) = 0.08314 <br>
s({Chocolate Cake},{Chocolate Coffee}) = 0.4404 <br>

The correlation coefficient is the same for both rules <br>{Chocolate Coffee} -> {Chocolate Cake} and {Chocolate Cake} -> {Chocolate Coffee} because the correlation coefficient is invariant. 

```python
intial_bakery_df = pd.read_csv("75000-out2-binary.csv")
intial_bakery_df.describe()
```

```python
intial_bakery_df[(intial_bakery_df["Chocolate Cake"] == 1) & (intial_bakery_df["Chocolate Coffee"] == 0)]["Transaction Number"].count()
```

```python
bake_f = apriori(intial_bakery_df.drop(columns="Transaction Number"), min_support = 0.04, use_colnames = True)
bake_f.sort_values('support', ascending = False)
```

```python
ccake_support = bake_f[bake_f["itemsets"] == {'Chocolate Cake'}]["support"].values[0]
print(ccake_support)
```

```python
ccoffee_support = bake_f[bake_f["itemsets"] == {'Chocolate Coffee'}]["support"].values[0]
print(ccoffee_support)
```

```python
pair_support = bake_f[bake_f["itemsets"] == {'Chocolate Coffee','Chocolate Cake'}]["support"].values[0]
print(pair_support)
```

```python
all_n = intial_bakery_df["Transaction Number"].count()
all_n
```

```python
fp1 = np.ceil(all_n*ccoffee_support)
fp1
```

```python
f1p = np.ceil(all_n*ccake_support)
f1p
```

```python
f11 = np.ceil(all_n*pair_support)
f11
```

```python
f01 = fp1 - f11
f01
```

```python
f10 = f1p - f11
f10
```

```python
f0p = all_n - f1p
f0p
```

```python
fp0 = all_n - fp1
fp0
```

```python
f00 = fp0 - f10 
f00
```

```python
f0p - f01 == f00
```

```python
f00 + f01 == f0p
```

```python
f00 + f10 == fp0
```

```python
f11 + f10 == f1p
```

```python
f11 + f01 == fp1
```

```python
print("\t","Coco","\t\t","!Coco","\t")
print()
print("Cake", "\t",f11,"\t",f10,"\t",f1p)
print()
print("!Cake","\t",f01,"\t",f00,"\t",f0p)
print()
print("\t", fp1, "\t", fp0, "\t", all_n)
```

```python
corr = (f11*f00 - f10*f01)/np.sqrt(fp1*f1p*fp0*f0p)
corr
```
